```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], kde=False, bins=100)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)#mean
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=max)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=sum)
```

```python
sns.countplot(x='sex',  data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,hue='smoker', palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, pallete='sets2')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', pallete='sets2')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',split=True, pallete='Set2')
```
